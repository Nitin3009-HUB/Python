```python
import pandas as pd
import numpy as np
```

```python
# Task 1: Create three dataframes and save them as .csv files
# Project DataFrame
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']}
project_df = pd.DataFrame(project_data)
project_df.to_csv('project_df.csv', index=False)  # Save to CSV
project_df
```

```python
# Employee DataFrame
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]}
employee_df = pd.DataFrame(employee_data)
employee_df.to_csv('employee_df.csv', index=False)  # Save to CSV
employee_df
```

```python
# Seniority Level DataFrame
seniority_data = {
    'ID' : ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level' : [2,2,3,2,3]}
seniority_df = pd.DataFrame(seniority_data)
seniority_df.to_csv('seniority_df.csv',index=False) # Save to CSV
seniority_df
```

```python
# Task 2: Compute missing values in 'Cost' column of 'Project' DataFrame and replace them with the running average using a "For" loop.
# Load the project_df from the csv file created in task 1
project_df = pd.read_csv('project_df.csv')
cost_values = []
for cost in project_df['Cost']:
    if pd.isna(cost):
        if cost_values:  # Check if cost_values list is not empty
            running_average = sum(cost_values) / len(cost_values)
            cost_values.append(running_average)
        else:
            cost_values.append(0)
    else:
        cost_values.append(cost)

project_df['Cost'] = cost_values

project_df
```

```python
# Task 3: Split the 'Name' column in the 'Employee' DataFrame into 'First Name' and 'Last Name' and remove the original 'Name' column.
# Load the employee_df from the csv file created in task 1
employee_df = pd.read_csv('employee_df.csv')
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', expand=True)
employee_df = employee_df.drop(columns=['Name'])
employee_df
```

```python
# Task 4: Join all three dataframes ('Project', 'Employee', 'Seniority Level') into a single dataframe named 'Final'.
final_df = pd.merge(pd.merge(project_df, employee_df, on='ID'), seniority_df, on='ID')
final_df
```

```python
# Task 5: Add a new 'Bonus' column to the 'Final' DataFrame. Give a 5% bonus on 'Project Cost' only to employees who have finished their projects.
final_df['Bonus'] = np.where(final_df['Status'] == 'Finished', final_df['Cost'] * 0.05, 0)
final_df
```

```python
# Task 6: Demote the designation level by 1 for projects with 'Failed' status. Delete employee records where 'Designation Level' is above 4.
failed_ids = final_df.loc[final_df["Status"] == "Failed", "ID"].unique()
mask = final_df["ID"].isin(failed_ids) & (final_df["Status"] == "Failed")
final_df.loc[mask, "Designation Level"] -= 1
final_df["Designation Level"] = final_df["Designation Level"].clip(lower=1)
final_df = final_df[final_df["Designation Level"] <= 4]
final_df


```

```python
# Task 7: Add "Mr." or "Mrs." to the 'First Name' column based on gender and drop the 'Gender' column.
def add_title(row):
    if row['Gender'] == 'M':
        return 'Mr. ' + row['First Name']
    elif row['Gender'] == 'F':
        return 'Mrs. ' + row['First Name']
    else:
        return row['First Name'] # Handle cases where gender is not 'M' or 'F'

final_df['First Name'] = final_df.apply(add_title, axis=1)
final_df = final_df.drop(columns=['Gender'])
final_df
```

```python
# Task 8: Promote 'Designation Level' by 1 for employees older than 29 using an IF condition.
final_df['Designation Level'] = np.where(final_df['Age'] > 29, final_df['Designation Level'] + 1, final_df['Designation Level'])
final_df
```

```python
# Task 9: Add the cost of all projects for each employee and save it in a new dataframe 'TotalProjCost' with columns 'ID', 'First Name', and 'Total Cost'.
TotalProjCost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index(name='Total Cost')
TotalProjCost

```

```python
# Task 10: Print all employee details whose city name contains the letter "o".
employees_with_o_in_city = final_df[final_df['City'].str.contains('o', case=False)] # case=False to include both 'o' and 'O'
employees_with_o_in_city
```
